# Análese de Dados de Filmes Exibidos no Brasil de 2009 a 2018
Com base nos dados da Ancine (http://oca.ancine.gov.br) vamos analisar os filmes exibidos no Brasil

Arquivo -> V2120.xlsx

```python
# Importando as bibliotecas
import pandas as pd
# Configurando o matplotlib
%matplotlib inline
```

```python
# Cria o DataFrame com base no arquivo Excel
# Precisamos pular as 2 primeiras Linhas (skiprows)
df = pd.read_excel('2120.xlsx', skiprows=2, skipfooter=16)
# Exibindo as 5 primeiras linhas do DataFrame
df.head()
```

```python
# Exibir as 5 ultimas linhas
df.tail(20)
```

```python
# Verificar todos os tipos das colunas
df.dtypes
```

```python
# Converter o campo Data de Lançamento em data
df['Data de lançamento'] = pd.to_datetime(df['Data de lançamento'])
```

```python
# vamos indetificar o problema da coluna Data de Lançamento
df['Data de lançamento'].value_counts()
```

```python
# Exibir alguns filmes que estão com data = Relançamento
df[df['Data de lançamento'] == 'Relançamento'].head(5)
```
